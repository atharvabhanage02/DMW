```python
from matplotlib import pyplot as plt
import numpy as np 
```

```python
x = np.arange(1,10,0.1)
y = 2*x+5
plt.plot(x,y,color='red')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x = np.arange(1,10,0.1)
y1 = 2*x+5
y2 = 3*x+10

plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

```python
# bar-plot
fruits = {'apple': 30 , 'mango':45 , 'banana':10}
names = list(fruits.keys())
quantity = list(fruits.values())

```

```python
names,quantity
```

```python
plt.bar(names,quantity)
plt.show()
```

```python
# Customizing Bar plot
plt.bar(names,quantity)
plt.title('Distribution of Fruits')
plt.xlabel('Fruits')
plt.ylabel("Quantity")
plt.show()
```

```python
# Customizing Bar plot (Horizontal)
plt.barh(names,quantity)
plt.title('Distribution of Fruits')
plt.xlabel('Fruits')
plt.ylabel("Quantity")
plt.show()
```

```python
# Scatter-plot
x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,4,3,2,9]
b = [5,6,7,8,3,2,1,4,9]

plt.scatter(x,a)
plt.scatter(x,b)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')

plt.show()
```

```python
# Histogram 
data = [1,1,1,1,4,6,3,2,4,4,4,4.2,7,8,4,5,3,2,1,7,8,9]
plt.hist(data)
plt.show()
```

```python
# Box-Plot

one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data = list([one,two,three]) 
plt.boxplot(data)
plt.show()
```

```python
# Violin-Plot
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data = list([one,two,three]) 
plt.violinplot(data,showmedians=True,showmeans=True)
plt.grid(True)
plt.title("Distribution of Data")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Pie-chart
players = ['Sachin', 'Virat' , 'Hardik' , 'Rohit' , 'SuryaKumar']
runs = [98,60,35,40,30]
plt.pie(runs,labels=players,autopct='%0.1f%%',explode = (0.1,0,0,0,0))
plt.show()
```

```python
# Area-plot
x= [10,20,30,40,50,60,70,80,90]
y = [1,1,2,5,6,4,7,4,6]

plt.stackplot(x,y)
plt.show()
```
